# Entendiendo el archivo .env y obteniendo tu clave API del LLM

## ¿Qué es un archivo .env?

Un archivo .env es un archivo de texto simple que almacena **variables de entorno** para tu proyecto. Piensa en él como un lugar de almacenamiento seguro para información sensible como claves API, contraseñas y configuraciones que no quieres incluir directamente en tus scripts de Python o compartir accidentalmente con otros.

El nombre ".env" literalmente significa "punto env" (el punto lo convierte en un archivo oculto en sistemas Unix), y sigue un formato simple: cada línea contiene un nombre de variable, un signo igual y su valor.

## ¿Por qué usar un archivo .env?

Imagina que estás escribiendo código de LangChain que usa la API de OpenAI. Podrías escribir tu clave API directamente en tu código así:

```python
openai_api_key = "sk-1234567890abcdef"  # ¡NO HAGAS ESTO!
```

Pero esto es problemático porque si compartes tu código en GitHub o con compañeros, estarás compartiendo tu clave API privada, y cualquiera podría usarla (y potencialmente generar cargos en tu cuenta). Un archivo .env soluciona esto manteniendo los secretos separados de tu código.

## Cómo crear un archivo .env

Crear un archivo .env es sencillo:

1. **Abre un editor de texto** (como el Bloc de notas, VS Code o cualquier editor de código)
2. **Crea un nuevo archivo** y guárdalo como `.env` (literalmente solo esos cuatro caracteres - un punto seguido de "env", sin extensión de archivo como .txt)
3. **Añade tus variables** en el formato CLAVE=VALOR
4. **Guarda el archivo** en el directorio raíz de tu proyecto

**Importante:** Asegúrate de que tu archivo .env esté listado en tu archivo `.gitignore` para que no se suba a GitHub. La mayoría de proyectos Python ya incluyen `.env` en el gitignore por defecto.

## ¿Por qué el directorio raíz?

El directorio raíz es la carpeta principal de tu proyecto (la carpeta de nivel superior que contiene todos tus archivos de ejercicio y subcarpetas). Cuando ejecutas `load_dotenv()`, la biblioteca busca el archivo .env comenzando desde el directorio donde se encuentra tu script Python, luego sube por los directorios padre hasta encontrar uno.

Al colocar el archivo .env en el directorio raíz, todos tus scripts de ejercicio (ya sea en subcarpetas o en el nivel superior) pueden acceder a las mismas variables de entorno. Esto significa que solo necesitas **un** archivo .env para todo tu curso, no uno separado para cada ejercicio.

## Cómo funciona load_dotenv()

Cuando escribes estas dos líneas al principio de tu ejercicio de LangChain:

```python
from dotenv import load_dotenv
load_dotenv()
```

Esto es lo que sucede paso a paso:

1. **Importar la función:** La primera línea importa la función `load_dotenv` del paquete `python-dotenv` (necesitas instalar este paquete primero con `pip install python-dotenv`)

2. **Cargar las variables:** La segunda línea `load_dotenv()` busca tu archivo .env, lo lee y carga todas las variables en el entorno de tu programa

3. **Acceder a las variables:** Después de cargar, puedes acceder a estas variables en tu código usando `os.getenv()` o están automáticamente disponibles para bibliotecas como LangChain que las buscan

Por ejemplo, después de ejecutar `load_dotenv()`, cuando inicializas un modelo LangChain, automáticamente encuentra tu clave API:

```python
from langchain_openai import ChatOpenAI

# LangChain automáticamente busca OPENAI_API_KEY en el entorno
llm = ChatOpenAI()  # ¡No es necesario pasar manualmente la clave API!
```

## Entendiendo tu archivo .env de ejemplo

Desglosemos cada sección del .env.example que incluimos en el directorio raíz del proyecto:

**Claves API de modelos LLM:**
```
OPENAI_API_KEY='tu_clave_api_openai_aqui'
ANTHROPIC_API_KEY='tu_clave_api_anthropic_aqui'
GOOGLE_API_KEY='tu_clave_api_google_aqui'
```

Estas almacenan las claves de autenticación para diferentes modelos de lenguaje IA. Obtendrías estas claves registrándote en el sitio web de cada proveedor (OpenAI, Anthropic, Google). Reemplaza el texto de ejemplo con tus claves reales, que normalmente parecen cadenas largas y aleatorias de letras y números.

**Herramientas de terceros:**
```
TAVILY_API_KEY='tu_clave_api_tavily_aqui'
```

Esta es para Tavily, una API de búsqueda que LangChain puede usar para buscar en la web. Solo la necesitarías si tus ejercicios implican dar a tu agente IA la capacidad de buscar en internet.

**Configuración de LangSmith:**
```
LANGSMITH_API_KEY='tu_clave_api_langsmith_aqui'
LANGSMITH_TRACING=true
LANGSMITH_PROJECT='tu_nombre_proyecto_aqui'
```

LangSmith es la plataforma de monitorización y depuración de LangChain. Estas variables configuran el seguimiento para tus aplicaciones LangChain. La configuración `TRACING=true` habilita la monitorización, y `PROJECT` nombra tu proyecto para que puedas organizar tus trazas. Esto es opcional pero útil para aprender porque puedes ver exactamente qué está haciendo tu aplicación LangChain entre bastidores.

**Comentarios y configuraciones opcionales:**
Las líneas que comienzan con `#` son comentarios - son ignoradas por el programa y están ahí solo para ayudarte a entender para qué sirve cada sección. La línea comentada `LANGSMITH_ENDPOINT` solo se descomentaría si estás usando los servidores europeos de LangChain en lugar de los predeterminados de EE.UU.

## Cómo obtener tu clave API de OpenAI

Dado que **los LLM de OpenAI son actualmente los más utilizados en el mundo empresarial**, usaremos los LLM de OpenAI en la mayoría de los ejercicios y proyectos de este bootcamp.

**Te mostraremos cómo usar LLM alternativos**, para que puedas experimentar con ellos por ti mismo.

Aquí tienes una guía sencilla para obtener tu clave API de OpenAI:

**Paso 1: Crear una cuenta de OpenAI**
Ve a https://platform.openai.com y haz clic en "Sign up" (Registrarse). Puedes usar tu dirección de correo electrónico o registrarte con Google/Microsoft.

**Paso 2: Añadir un método de pago**
Una vez que hayas iniciado sesión, necesitarás añadir un método de pago (tarjeta de crédito/débito). OpenAI cobra según el uso, pero no te preocupes - para ejercicios de aprendizaje, los costes son típicamente muy pequeños (a menudo solo unos céntimos por sesión).

**Paso 3: Obtener tu clave API**
- Haz clic en el icono de tu perfil en la esquina superior derecha
- Selecciona "API keys" (Claves API) del menú desplegable
- Haz clic en el botón "+ Create new secret key" (Crear nueva clave secreta)
- Dale un nombre (como "LangChain Course" o "Curso LangChain")
- Copia la clave inmediatamente y guárdala en un lugar seguro (como un gestor de contraseñas o nota segura)

**Consejos importantes:**
- **Nunca compartas tu clave API** - trátala como una contraseña
- **Guárdala inmediatamente** - OpenAI solo la muestra una vez. Si la pierdes, necesitarás crear una nueva
- **Establece límites de gasto** - En la configuración de tu cuenta, puedes establecer un límite de gasto mensual para evitar cargos inesperados

**Cómo usarás tu clave API durante el curso:**
Guardarás tu clave API en un archivo `.env` en el mismo directorio raíz del ejercicio en el que estés trabajando.

## Cómo obtener tu clave API de Anthropic
Aquí tienes una guía sencilla para obtener tu clave API de Anthropic:

**Paso 1: Crear una cuenta de Anthropic**
Ve a https://console.anthropic.com y haz clic en "Sign up" (Registrarse). Puedes crear una cuenta usando tu dirección de correo electrónico.

**Paso 2: Añadir un método de pago**
Después de iniciar sesión, necesitarás añadir un método de pago (tarjeta de crédito/débito). Al igual que OpenAI, Anthropic cobra según el uso, y para ejercicios de aprendizaje, los costes son típicamente muy bajos.

**Paso 3: Obtener tu clave API**
- Una vez conectado a la Consola, busca "API Keys" (Claves API) en el menú de la barra lateral izquierda
- Haz clic en "+ Create Key" (Crear clave)
- Dale un nombre descriptivo (como "LangChain Course" o "Curso LangChain")
- Copia la clave inmediatamente y guárdala en un lugar seguro (gestor de contraseñas o nota segura)

**Consejos importantes:**
- **Nunca compartas tu clave API** - es como una contraseña que da acceso a tu cuenta
- **Guárdala de inmediato** - Solo puedes ver la clave completa una vez. Si la pierdes, necesitarás crear una nueva
- **Monitoriza el uso** - Consulta tu panel de uso para hacer seguimiento del gasto
- **Establece límites de presupuesto** - Puedes configurar límites de gasto en la configuración de tu cuenta para controlar los costes

**Cómo usarás tu clave API durante el curso:**
Guardarás tu clave API en un archivo `.env` en el mismo directorio raíz del ejercicio en el que estés trabajando.

## Cómo obtener tu clave API de Google para usar LLM de Google (Gemini, etc)
Aquí tienes una guía sencilla para obtener tu clave API de Google para Gemini:

**Paso 1: Ve a Google AI Studio**
Visita https://aistudio.google.com e inicia sesión con tu cuenta de Google (Gmail).

**Paso 2: Acepta los términos y condiciones**
Necesitarás aceptar los términos de servicio de Google para usar sus servicios de IA.

**Paso 3: Obtén tu clave API**
- Haz clic en "Get API key" (Obtener clave API) en la barra lateral izquierda (o podrías verlo de forma prominente en la página principal)
- Haz clic en "Create API key" (Crear clave API)
- Elige crear un nuevo proyecto o selecciona un proyecto existente de Google Cloud
- Tu clave API se generará - cópiala inmediatamente y guárdala de forma segura

**Consejos importantes:**
- **Mantenla privada** - Nunca compartas tu clave API ni la incluyas en repositorios de código públicos
- **Guárdala inmediatamente** - Almacénala en un gestor de contraseñas o nota segura
- **Nivel gratuito disponible** - Gemini ofrece un generoso nivel gratuito para aprender y experimentar, lo cual es genial para ejercicios del curso
- **Límites de tasa** - El nivel gratuito tiene límites de tasa, pero normalmente son más que suficientes para propósitos de aprendizaje
- **Monitoriza el uso** - Puedes hacer seguimiento del uso de tu API en Google AI Studio

**Para tu curso:**
Cuando tus ejercicios de LangChain necesiten usar modelos Gemini, típicamente guardarás la clave API como una variable de entorno llamada `GOOGLE_API_KEY` en tu archivo `.env`, junto con tus otras claves API.

**Cómo usarás tu clave API durante el curso:**
Guardarás tu clave API en un archivo `.env` en el mismo directorio raíz del ejercicio en el que estés trabajando.

## Cómo experimentar con otros LLM populares (código cerrado y código abierto)

A medida que aprendes LangChain y otros frameworks de IA Generativa, será valioso experimentar con diferentes Modelos de Lenguaje Grande (LLM) más allá de solo OpenAI, Anthropic y Google. Cada modelo tiene fortalezas, precios y capacidades únicos que pueden mejorar tu experiencia de aprendizaje. Exploremos algunas opciones populares.

#### Entendiendo código abierto vs. código cerrado

**LLM de código cerrado**: Estos son modelos propietarios a los que accedes a través de APIs. Pagas por uso, pero la empresa mantiene y mejora el modelo por ti.

**LLM de código abierto**: Los pesos del modelo están disponibles públicamente. Puedes ejecutarlos localmente (si tienes hardware potente) o usarlos a través de servicios API. A menudo son más asequibles y personalizables.

#### LLM populares de código cerrado para explorar

1. **DeepSeek** (China)
**Qué lo hace especial**: Excelente en tareas de programación y matemáticas, a menudo a una fracción del coste de otros modelos premium.

**Dónde empezar**:
- Sitio web: https://platform.deepseek.com
- Regístrate con correo electrónico
- Navega a la sección de claves API
- Precio muy asequible, genial para estudiantes con presupuesto limitado

**Mejor para**: Ejercicios de programación, resolución de problemas técnicos

---

2. **Grok** (xAI - empresa de Elon Musk)
**Qué lo hace especial**: Tiene acceso en tiempo real a datos de X/Twitter, conocido por un tono más conversacional y a veces humorístico.

**Dónde empezar**:
- Sitio web: https://x.ai (actualmente disponibilidad limitada)
- El acceso API se está expandiendo gradualmente
- Consulta su plataforma de desarrolladores para actualizaciones

**Mejor para**: Proyectos que involucran datos de redes sociales, aplicaciones conversacionales

---

3. **Kimi** (Moonshot AI - China)
**Qué lo hace especial**: Ventana de contexto largo excepcional (puede manejar documentos muy largos), fuertes capacidades multilingües.

**Dónde empezar**:
- Sitio web: https://platform.moonshot.cn
- Interfaz principalmente en chino, pero soporta inglés
- Crea una cuenta y navega a la sección API

**Mejor para**: Procesamiento de documentos largos, proyectos multilingües

---

4. **Cohere**
**Qué lo hace especial**: Enfocado a empresas con fuertes capacidades de incrustación y clasificación de texto, excelente documentación para desarrolladores.

**Dónde empezar**:
- Sitio web: https://cohere.com
- Haz clic en "Get Started" (Comenzar) o "Sign Up" (Registrarse)
- Panel → Claves API
- Generoso nivel de prueba gratuito para aprender

**Mejor para**: Clasificación de texto, búsqueda semántica, incrustaciones

---

#### LLM populares de código abierto para explorar

5. **Mistral** (Francia)
**Qué lo hace especial**: Alternativa europea de alta calidad, varios tamaños de modelo, fuerte relación rendimiento-coste.

**Dónde empezar**:
- Sitio web: https://console.mistral.ai
- Crear cuenta
- Sección de claves API
- También ofrece modelos de código abierto que puedes descargar

**Mejor para**: Necesidades de residencia de datos europea, experimentación rentable

---

6. **Modelos Meta Llama**
**Qué los hace especiales**: Una de las familias de modelos de código abierto más populares, en constante mejora, gran apoyo de la comunidad.

**Cómo acceder**:
- **A través de APIs**: Muchos proveedores alojan Llama (Together AI, Replicate, Groq)
- **Together AI**: https://together.ai - Fácil acceso API
- **Replicate**: https://replicate.com - Pago por uso
- **Groq**: https://groq.com - Inferencia ultra rápida
- **Ejecutar localmente**: Descargar de Hugging Face (requiere hardware potente)

**Mejor para**: Aprender sobre modelos de código abierto, proyectos comunitarios

---

7. **Hugging Face** (Plataforma)
**Qué lo hace especial**: No es un modelo único, sino una plataforma que aloja miles de modelos de código abierto. Piensa en ello como el GitHub de los modelos de IA.

**Dónde empezar**:
- Sitio web: https://huggingface.co
- Crear cuenta gratuita
- Configuración → Tokens de acceso
- Acceder a miles de modelos a través de su API

**Mejor para**: Experimentar con muchos modelos diferentes, aprender sobre variedades de modelos

---

#### Resumen comparativo rápido

**DeepSeek** es un modelo de código cerrado con coste muy bajo, excelente para programación y matemáticas. Muy recomendado para principiantes que buscan gran valor.

**Grok** es un modelo de código cerrado con precio medio que sobresale en acceso a datos en tiempo real. Sin embargo, actualmente tiene disponibilidad limitada, por lo que puede no ser el mejor punto de partida para principiantes.

**Kimi** es un modelo de código cerrado con precio medio, excepcional en manejar contexto largo. La interfaz principalmente china podría presentar una curva de aprendizaje para algunos estudiantes.

**Cohere** es un modelo de código cerrado que ofrece un generoso nivel gratuito y excelente documentación. Se especializa en incrustaciones y es muy amigable para principiantes.

**Mistral** es un modelo de código abierto con bajo coste y APIs fáciles de usar. Es una gran alternativa europea y recomendada para principiantes.

**Llama** (Meta) es un modelo de código abierto que es gratis o de bajo coste dependiendo de cómo accedas a él. Con apoyo masivo de la comunidad y amplia popularidad, es excelente para principiantes que quieren explorar opciones de código abierto.

**Hugging Face** es una plataforma (no un modelo único) con costes variables dependiendo del modelo que uses. Es altamente educativa y perfecta para principiantes que quieren explorar la variedad de modelos disponibles.

#### Consejos para experimentar con múltiples LLM

**Empieza con niveles gratuitos**: La mayoría de servicios ofrecen créditos o pruebas gratuitas - aprovéchalos para aprender.

**Usa variables de entorno**: Almacena todas tus claves API en tu archivo `.env`:
```
OPENAI_API_KEY=tu_clave_aqui
ANTHROPIC_API_KEY=tu_clave_aqui
DEEPSEEK_API_KEY=tu_clave_aqui
MISTRAL_API_KEY=tu_clave_aqui
```

**Compara salidas**: Ejecuta el mismo prompt a través de diferentes modelos para ver cómo difieren en respuestas, velocidad y coste.

**Establece límites de gasto**: Siempre configura alertas de presupuesto en plataformas que las soporten.

**Lee la documentación**: Cada LLM tiene diferentes capacidades - consulta sus documentos para entender qué hacen mejor.

**Únete a comunidades**: Muchos de estos modelos tienen servidores Discord o foros donde puedes aprender de otros usuarios.

---

#### ¿Por qué experimentar con diferentes modelos?

**Aprende flexibilidad**: En proyectos del mundo real, a menudo necesitarás elegir el modelo correcto para tareas específicas.

**Optimización de costes**: Algunos modelos son mucho más baratos para ciertas tareas.

**Entiende las compensaciones**: Aprenderás sobre el equilibrio entre velocidad vs. calidad, y coste vs. capacidad.

**Evita el bloqueo de proveedor**: No dependas de un solo proveedor - diversifica tus habilidades.

**Amplía tu conjunto de habilidades**: Cada API tiene ligeras diferencias - aprender múltiples te hace más versátil como desarrollador.


Recuerda: no necesitas registrarte en todo a la vez. Empieza con 2-3 modelos y expande a medida que te sientas más cómodo con LangChain. ¡Cada nuevo modelo que aprendes añade otra herramienta a tu kit de desarrollo de IA!
